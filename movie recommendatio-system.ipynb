```python
import ast
```

```python
def convert(obj):
   L=[]
   for i in ast.literal_eval(obj):
         L.append(i['name']) 
   return L
    
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data.head()
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
data.head()
```

```python
def convert2(obj):
   L=[]
   count=0
   for i in ast.literal_eval(obj):
       if count !=3:
         L.append(i['name']) 
         count+=1
       else:
          break
       
   return L
```

```python
data['cast']=data['cast'].apply(convert)
```

```python
def fetch_dir(obj):
   L=[]

   for i in ast.literal_eval(obj):
       if i['job']=='Director':
         L.append(i['name']) 
         break
       
       
   return L
```

```python
data['crew']=data['crew'].apply(fetch_dir)
data.head()
```

```python
data['overview'][0]
```

```python
data['overview']=data['overview'].apply(lambda x:x.split())
```

```python
data.head()
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast']=data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head()
```

```python
data['tags']=data['overview']+data['keywords']+data['genres']+data['cast']+data['crew']
```

```python
data=data.drop(columns=['overview','keywords','genres','cast','crew'])
```

```python
data.head()
```

```python
data['tags']=data['tags'].apply(lambda x:" ".join(x))
```

```python
data['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(min_df=3,max_features=None)
```

```python
X=vectorizer.fit_transform(data['tags']) 
X.shape
```

```python
print(X)
```

```python
similar=cosine_similarity(X)
```

```python
print(similar)
```

```python
user_inp=input("Enter the movie you want to see")
```

```python
print(similar)
```

```python
user_inp=input("Enter the movie you want to see:")
```

```python
all_titles=data['title'].tolist()
```

```python
find_close_match=get_close_matches(user_inp,all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_movie=data[data.title==close_match]['id'].values[0]
print(index_of_movie)
```

```python
similarity_score=list(enumerate(similar[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score ,key= lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
print("Movies suggested for you:\n")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=data[data.index==index]['title'].values[0]
    if (i<11):
        print(i, '.' , title_from_index)
        i+=1
```
